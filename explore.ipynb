```python
from pokertraining.deck import Deck
from pokertraining.table import Table
from pokertraining.player import Player
```

```python
# Initialize table

n = 9
t = Table(n)
players = [t.addPlayer(Player(f'abc{i}')) for i,p in enumerate(range(n))]
```

```python
# Get Dealer
t.init_dealer();
```

```python
def playhand(t):
    t.getDeckAndShuffle()
    t.burnCard()
    t.dealHoleCards()

    t.burnCard()
    t.dealCommunityCards('flop')
    t.burnCard()
    t.dealCommunityCards('turn')
    t.burnCard()
    t.dealCommunityCards('river')

    t.getBestHands('river')
    t.getWinners()
    return t
```

```python
import json
import os
t = playhand(t)

hands = []
for player in t.players.values():
    hands.append({
        'playerCount':len(t.players),
        'holeCards':[c.toJson() for c in player.showHand()],
        'flop': [c.toJson() for c in t.flop],
        'turn': t.turn.toJson(),
        'river': t.river.toJson(),
        'win': player.win
    })

fname = "handLog.json"
if os.path.isfile(fname):
    # File exists
    with open(fname, 'a+') as outfile:
        outfile.seek(0, os.SEEK_END)              # seek to end of file; f.seek(0, 2) is legal
        outfile.seek(outfile.tell()-1, os.SEEK_SET)
        outfile.truncate()
        outfile.write(',\n')
        [json.dump(hand, outfile) for hand in hands]
        outfile.write(']')
else: 
    # Create file
    with open(fname, 'w') as outfile:
        array = []
        [array.append(hand) for hand in hands]
        json.dump(array, outfile)
        outfile.write('\n')

t.clearCommunityCards()
for player in t.players.values():
    player.clearHand()
```

```python
with open('handLog.json', 'r',encoding='utf-8') as log:
    hands = json.load(log)
```
